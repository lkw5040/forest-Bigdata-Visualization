### 후보도시 32곳-> 21곳 94건

```python
price
```

```python
price = price.reset_index()
price = price.drop(['index'], axis = 1)
```

```python
change = ['고양', '안산', '경산']

price['후보지'] = price['시도명'] + price['시군구명']

price['후보지'] = np.where(price.시군구명.str.contains('|'.join(change)),
                           price['후보지'],
                           price['후보지']+ price.소재지.str.split(' ').str[0])

price
```

```python
price = price[['후보지', '면적', '공시지가']]
```

```python
price
```

```python
area = pd.DataFrame(price.groupby(['후보지'])['면적'].agg('sum'))
area = area.reset_index()
area.head()
```

```python
official_price = pd.DataFrame(price.groupby(['후보지'])['공시지가'].agg('mean'))
official_price = official_price.reset_index()
official_price.head()
```

```python
candidate_list = [i.replace(' ', '') for i in candidate_list]
```

```python
df = pd.DataFrame(candidate_list, columns=['후보지'])
```

```python
merge = pd.merge(df, area, how='left', left_on='후보지', right_on='후보지')
merge = pd.merge(merge, official_price, how='left', left_on='후보지', right_on='후보지')
merge = merge.fillna(0)
merge['면적'] = merge['면적'].astype(int)
merge['공시지가'] = merge['공시지가'].astype(int)

merge
```

```python
def normalize(arr, t_min, t_max):
    norm_arr=[]
    diff=t_max-t_min
    diff_arr=max(arr) - min(arr)
    for i in arr:
        temp=(((i-min(arr))*diff)/diff_arr)+t_min
        norm_arr.append(temp)
    return norm_arr
range_to_normalize=(0, 1)
```

```python
area = merge['면적']
normalized_area=normalize(area, range_to_normalize[0], range_to_normalize[1])
merge['norm_면적'] = normalized_area

price = merge['공시지가']
normalized_price = normalize(price, range_to_normalize[0], range_to_normalize[1])
merge['norm_공시지가'] = normalized_price
```
